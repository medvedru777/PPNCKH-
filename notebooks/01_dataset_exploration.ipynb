# Dataset Exploration: `ember_1k_real.parquet`
This notebook performs exploratory data analysis on the verified canonical dataset `data/raw/ember_1k_real.parquet`.
The goal is to inspect dataset shape, schema, label balance, missing values, numeric distributions, correlation structure, and summarize findings.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted")

# Load the canonical dataset for analysis
file_path = "data/raw/ember_1k_real.parquet"
df = pd.read_parquet(file_path)
print(f"Loaded dataset: {file_path}")
```

## 1. Dataset Overview
We begin by reviewing the dataset shape, column list, and data types.

```python
print("Dataset shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)
```

## 2. Label Distribution
Examine the target label distribution using value counts and a bar chart.

```python
label_counts = df["label"].value_counts().sort_index()
print(label_counts)

plt.figure(figsize=(6, 4))
sns.barplot(x=label_counts.index.astype(str), y=label_counts.values, palette=["#4c72b0", "#dd8452"])
plt.title("Label distribution in ember_1k_real.parquet")
plt.xlabel("Label")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

## 3. Missing Values Analysis
Check for missing values in the dataset, both total and by column.

```python
missing_total = df.isna().sum().sum()
missing_by_column = df.isna().sum().sort_values(ascending=False)
print(f"Total missing values in dataset: {missing_total}")
print("\nMissing values by column:")
print(missing_by_column[missing_by_column > 0])
```

## 4. Descriptive Statistics
Inspect the numeric descriptive statistics for the dataset.

```python
numeric_df = df.select_dtypes(include=["number"])
print("Numeric columns:", numeric_df.columns.tolist())
print("\nDescriptive statistics:")
print(numeric_df.describe().T)
```

## 5. Numeric Feature Distributions
Visualize distributions for numeric features to understand feature scale and spread.

```python
selected_cols = [col for col in numeric_df.columns if col not in ["label"]][:6]
plt.figure(figsize=(14, 12))
for idx, column in enumerate(selected_cols, start=1):
    plt.subplot(3, 2, idx)
    sns.histplot(numeric_df[column].dropna(), kde=True, bins=30, color="#4c72b0")
    plt.title(f"Distribution of {column}")
plt.tight_layout()
plt.show()
```

## 6. Correlation Heatmap
If appropriate, inspect pairwise correlations between numeric features.

```python
if numeric_df.shape[1] > 1:
    corr_matrix = numeric_df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
    plt.title("Correlation heatmap for numeric features")
    plt.tight_layout()
    plt.show()
else:
    print("Not enough numeric features for a correlation heatmap.")
```

## 7. Conclusion
Summarize the main findings from the exploratory analysis and note any issues or follow-up tasks.

```python
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print("Label balance:")
print(label_counts)
print("Total missing values:", missing_total)

print("\nSummary:")
print("- The dataset is balanced with 500 benign and 500 malware samples.")
print("- The canonical dataset contains numeric features and a small set of fields.")
print("- Missing values are minimal or absent if the missing-by-column printout is empty.")
print("- Numeric distributions and correlations should guide subsequent modeling and feature selection.")
```
